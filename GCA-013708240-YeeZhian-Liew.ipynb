## Credit German Risk

When a bank receives a loan application, based on the applicant’s profile the bank has to make a decision regarding whether to go ahead with the loan approval or not. Two types of risks are associated with the bank’s decision –

If the applicant is a good credit risk, i.e. is likely to repay the loan, then not approving the loan to the person results in a loss of business to the bank
If the applicant is a bad credit risk, i.e. is not likely to repay the loan, then approving the loan to the person results in a financial loss to the bank
Objective of Analysis:

Minimization of risk and maximization of profit on behalf of the bank.

To minimize loss from the bank’s perspective, the bank needs a decision rule regarding who to give approval of the loan and who not to. An applicant’s demographic and socio-economic profiles are considered by loan managers before a decision is taken regarding his/her loan application.

The German Credit Data contains data on 20 variables and the classification whether an applicant is considered a Good or a Bad credit risk for 1000 loan applicants.  

A predictive model developed on this data is expected to provide a bank manager guidance for making a decision whether to approve a loan to a prospective applicant based on his/her profiles.

We need to get the balance between sensitivity and specificity. 

When we decrease the threshold, we get more positive values therefore it increases the sensitivity and decreasing the specificity.

Similarly, when we increase the threshold, we get more negative values therefore we get higher specificity and lower sensitivity.

In this case, the default of the threhold is 0.5, we can decrease it to 0.4 to have a better model.
- Sensitivity increase from 0.47 to 0.61

## Learning curve 
Plot of the training/cross-validation error VS the sample size. 
The learning curve can be used to detect whether the model has the **high bias or high variance.** 

**High bias/underfitting**, as for the sample size increases, training error will increase and the cross validation error will decrease. They will be very close to each other but still at a high error rate for both training and classification error. 

**You should:**

**1) Try adding/creating more features**

**2) Try decreasing the regularisation parameter λ**

**High variance/overfitting**, as for the sample size increases, the training error will keep increasing and cross-validation error will keep decreasing. They will end up at a low training and cross-validation error rate. 



**You should:**

**1) Try increasing samples**


```python
import warnings
import matplotlib.cbook
warnings.filterwarnings("ignore",category=matplotlib.cbook.mplDeprecation)
from warnings import simplefilter
simplefilter(action='ignore', category=FutureWarning)
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures


for d in [2]:
    model = make_pipeline(PolynomialFeatures(degree=d), LogisticRegression())
    

from sklearn.model_selection import learning_curve
train_sizes, train_scores, validation_scores = learning_curve(model, X_train, y_train, train_sizes=np.logspace(-1, 0, 10))


plt.figure(figsize=(6, 4))
plt.plot(train_sizes, validation_scores.mean(axis=1),
            lw=2, label='cross-validation')
plt.plot(train_sizes, train_scores.mean(axis=1),
                lw=2, label='training')
plt.ylim(ymin=-.1, ymax=1)

plt.legend(loc='best')
plt.xlabel('number of train samples')
plt.ylabel('explained variance')
plt.title('Learning curve (degree=%i)' % d)
plt.tight_layout()

plt.show()
```

Seems like cross-validation and training curve are from diverging and getting converging when it reaches 250 samples. 

It shows our graph is having a high variance problem because there is a huge gap in between cross validation and training line. 

In this case, we can improve our model accuracy by increasing training samples.
